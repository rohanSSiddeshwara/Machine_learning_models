```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## build a linear regresion class
class LinearRegression:
    def __init__(self):
        self.w = None
        self.b = None
        self.loss = []
        self.w_hist = []
        self.b_hist = []

    def fit(self, X, y, epochs=100, lr=0.01):
        print(X.shape)
        self.w = np.random.randn(X.shape[1])
        self.b = np.random.randn()

        for i in range(epochs):
            y_hat = self.predict(X)
            print(y_hat.shape)
            loss = np.mean((y_hat - y)**2)
            self.loss.append(loss)

            dw = np.mean((y_hat - y) * X, axis=0)
            db = np.mean(y_hat - y)

            self.w -= lr * dw
            self.b -= lr * db

            self.w_hist.append(self.w)
            self.b_hist.append(self.b)

    def predict(self, X):
        return X * self.w + self.b

    def plot_loss(self):
        plt.plot(self.loss)
        plt.xlabel("epochs")
        plt.ylabel("loss")
        plt.show()

    ##plot y=mx+b curve
    def plot_regression(self, X, y):
        plt.scatter(X, y)
        plt.plot(X, self.predict(X))
        plt.show()
        
```

```python
## build a dataset
X = np.random.randn(100, 1)
y = 2 * X + 3 + np.random.randn(100, 1)

## train the model
model = LinearRegression()
model.fit(X, y, epochs=1000, lr=0.01)

## plot the loss
model.plot_loss()

## plot the regression
model.plot_regression(X, y)

## predict
model.predict(np.array([[1]]))

## save the model
import pickle
with open("model.pkl", "wb") as f:
    pickle.dump(model, f)

```

```python
# plot a regression line for all the values in whistory and b history
def plot_regression_line(w_hist, b_hist, X, y):
    for w, b in zip(w_hist, b_hist):
        plt.plot(X, w * X + b, color="red", alpha=0.1)
    plt.scatter(X, y)
    plt.show()

plot_regression_line(model.w_hist, model.b_hist, X, y)

```
